```python
from etl.generic_provider import GenericDataProvider

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
co2_daily_provider = GenericDataProvider("datalake/clean/ICAP/", want_metadata=False, tickers="CO2_EUA_PRICE")

co2_price_daily = co2_daily_provider.get_series("CO2_EUA_PRICE", freq="D")
co2_price_monthly = co2_daily_provider.get_series("CO2_EUA_PRICE", freq="D", resample_by="M", group_mode="mean")
```

## Daily aggregation

```python
ax = co2_price_daily.plot()
```

## Monthly aggregation

```python
ax = co2_price_monthly.plot()
ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.set_title("CO2 EUAs price (per month)")
```
