```python
from etl.generic_provider import GenericDataProvider

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
omie_provider = GenericDataProvider("datalake/clean/omie/", want_metadata=False, tickers="PRECIO_OMIE")

omie_price_monthly = omie_provider.get_series("PRECIO_OMIE", freq="H", resample_by="M", group_mode="mean")
omie_price_yearly = omie_provider.get_series("PRECIO_OMIE", freq="H", resample_by="Y", group_mode="mean")
```

## Energy SPOT price


### Monthly aggregation

```python
ax = omie_price_monthly.plot()
ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.set_title("SPOT market price (per month)")
```

```python
plot_acf(omie_price_monthly, lags=50);
plt.xlabel("Lag")
plt.title("SPOT market price (month) ACF plot")
```

### Yearly aggregation

```python
ax = omie_price_yearly.plot()
ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.set_title("SPOT market price (year)")
```

```python
plot_acf(omie_price_yearly, lags=25);
plt.xlabel("Lag")
plt.title("SPOT market price (year) ACF plot")
```
